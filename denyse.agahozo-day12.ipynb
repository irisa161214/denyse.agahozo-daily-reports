### Exercise 9:

```python
# Exercise 16
```

---

### Exercise 17:
Create a plot with multiple subplots in a 2x2 grid. Plot `y1 = x`, `y2 = x^2`, `y3 = x^3`, and `y4 = x^4`.

```python
# Exercise 17
```

---

````python
x= np.linspace(-10, 10, 100)
y1 = x
y2 = x**2
y3 = x**3
y4 = x**4
figure, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('Linear Function')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('y')

axs[0, 1].plot(x,y2,color='orange')
axs[0, 1].set_title('Quadratic Function')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('y')


axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('Cubic Function')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('y')
axs[1, 1].plot(x, y4, color='red')
axs[1, 1].set_title('Quartic Function')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('y')


# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
````

### Exercise 20:
Plot a sine wave using a dashed line and a cosine wave using a solid line on the same plot.

```python
# Exercise 20
```

---

````python
# 256 linearly spaced values between -pi and +pi
# both endpoints would be included; check (X[0], X[-1])
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C,color='green')
plt.plot(X, S,color='red',ls='--')
plt.show()
````
